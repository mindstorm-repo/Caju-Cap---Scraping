```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'http://cajucap.com.br/resultados-atualizado/'
```

```python
req = requests.get(url)
```

```python
if req.status_code == 200:
    content = req.content
elif req.status_code == 404:
    print('Página não encontrada.')
```

```python
soup = BeautifulSoup(content, 'html.parser')
```

```python
soup
```

```python
grupos = soup.find_all('div', class_='grupo')
```

```python
grupos
```

```python
len(grupos)
```

```python
grupos[1]
```

```python
premios = []
ticket = grupos[1].find_all('div', class_='sp')[1]
premio = {'titulo': grupos[1].h1.text, 'valor': grupos[1].h3.text}

nome = ticket.p
certificado = nome.find_next_sibling('p')
cidade = certificado.find_next_sibling('p')
vendedor = cidade.find_next_sibling('p')

ganhador = {'nome': nome.text, 'certificado': certificado.text, 'cidade': cidade.text, 'vendedor': vendedor.text}
premio['ganhador'] = ganhador

premio
```
